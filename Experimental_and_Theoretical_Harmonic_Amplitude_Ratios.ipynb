```python
import matplotlib.pyplot as plt
import numpy as np

# harmonic numbers from 1 to 10
n = np.array(range(1, 11))

# An/A1 (experimental)  ratios (using 'A_n/A_1 from Proc#4' is the experimental ratio)
A_n_A_1_exp = np.array([1, 0.999550005, 3.418819188, 0.054776039, 11.21182266, 0.084797891, 9.305699482, 0.101893096, 7.93442623, 0.13292011])

# 1/n^2 (theoretical)  
one_over_n_squared = 1 / n**2


plt.figure(figsize=(10, 6))
plt.plot(n, A_n_A_1_exp, 'o-', label='Experimental $A_n/A_1$')
plt.plot(n, one_over_n_squared, 's-', label='Theoretical $1/n^2$')
plt.title('Comparison of Experimental $A_n/A_1$ and Theoretical $1/n^2$')
plt.xlabel('Harmonic Number ($n$)')
plt.ylabel('Amplitude Ratio / Theoretical Value')
plt.grid(True)
plt.legend()
plt.savefig("plot_6.png", dpi=500)
plt.show()
```
