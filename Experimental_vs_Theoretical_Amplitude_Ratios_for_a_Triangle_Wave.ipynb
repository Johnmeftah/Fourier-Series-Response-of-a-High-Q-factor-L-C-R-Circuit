```python
import matplotlib.pyplot as plt
import numpy as np

# harmonic numbers
n = np.array([1, 2, 3, 4, 5, 6])

#  An/A1 (experimental)
An_A1_exp = np.array([1, 0.015375339, 0.126318963, 0.015676816, 0.055471812, 0.015676816])

#  An/A1  (theoretical) for a triangle wave (only odd harmonics, inversely proportional to n^2)
An_A1_theory = np.array([1, 0, 0.111111111, 0, 0.04, 0])


plt.figure(figsize=(10, 6))
plt.plot(n, An_A1_exp, 'o-', label='Experimental An/A1')
plt.plot(n, An_A1_theory, 's-', label='Theoretical An/A1 (Triangle Wave)')
plt.title('An/A1 vs n for Triangle Wave')
plt.xlabel('Harmonic number n')
plt.ylabel('Amplitude ratio An/A1')
plt.grid(True)
plt.legend()
plt.savefig("plot_5.png", dpi=500)
plt.show()
```
